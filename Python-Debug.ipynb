```python
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
data = sio.loadmat('./src/data_from_python/image_image_conv.mat')
```

```python
data.keys()
```

```python
kernel = data['kernel_image_image_conv']
inp = data['input_image_image_conv']
```

```python
nc = 3
mk = 5
nk = 5
shit = np.zeros((mk,nk))

plt.imshow(inp)
plt.show()

plt.imshow(100.0 * kernel)
plt.show()


for i in range(nc):
    shit = shit + signal.convolve2d(inp[:,:,i], kernel[:,:,i], mode='valid')
print(shit)
```

```python
nc = 3
mk = 5
nk = 5
shit_3d_sum = np.zeros((nk, mk,nk))

for i in range(nc):
    shit_3d_sum[i,:,:] = shit_3d_sum[i,:,:] + signal.convolve2d(inp[:,:,i], kernel[:,:,i], mode='valid')
shit_3d_sum = np.sum(shit_3d_sum,axis=0)
print(shit_3d_sum)
```
